Aplikasi Toko Sepatu

```python
# Data awal sepatu dan transaksi
stok_sepatu = {
    "Nike Air Max": {"stok": 10, "harga": 1500000},
    "Adidas Ultraboost": {"stok": 8, "harga": 1800000},
    "Puma RS-X": {"stok": 5, "harga": 1300000}
}
riwayat_pembelian = []
data_pembeli = []

# Fungsi untuk penjual
def menu_penjual():
    pemasukan = sum(item['total_harga'] for item in riwayat_pembelian)
    while True:
        print("\n--- Menu Penjual ---")
        print("1. Lihat Pemasukan")
        print("2. Tambah Sepatu ke Stok")
        print("3. Lihat Stok Sepatu")
        print("4. Lihat Riwayat Pembelian")
        print("5. Lihat Data Pembeli")
        print("0. Kembali ke Menu Utama")

        pilihan = input("Pilih menu: ")
        if pilihan == "1":
            print(f"Pemasukan total: Rp{pemasukan:,}")
        elif pilihan == "2":
            nama = input("Nama sepatu: ")
            jumlah = int(input("Jumlah stok: "))
            harga = int(input("Harga satuan: "))
            if nama in stok_sepatu:
                stok_sepatu[nama]['stok'] += jumlah
            else:
                stok_sepatu[nama] = {"stok": jumlah, "harga": harga}
            print("Stok berhasil ditambahkan.")
        elif pilihan == "3":
            print("--- Daftar Stok Sepatu ---")
            for nama, info in stok_sepatu.items():
                print(f"{nama} - Stok: {info['stok']} - Harga: Rp{info['harga']:,}")
        elif pilihan == "4":
            print("--- Riwayat Pembelian ---")
            for r in riwayat_pembelian:
                print(f"{r['nama_pembeli']} membeli {r['nama_sepatu']} sebanyak {r['jumlah']} - Total: Rp{r['total_harga']:,}")
        elif pilihan == "5":
            print("--- Data Pembeli ---")
            for d in data_pembeli:
                print(f"Nama: {d}")
        elif pilihan == "0":
            break
        else:
            print("Pilihan tidak valid.")

# Fungsi untuk pembeli
def menu_pembeli():
    while True:
        print("\n--- Menu Pembeli ---")
        print("1. Lihat Stok Sepatu")
        print("2. Pesan Sepatu")
        print("0. Kembali ke Menu Utama")

        pilihan = input("Pilih menu: ")
        if pilihan == "1":
            print("--- Daftar Stok Sepatu ---")
            for nama, info in stok_sepatu.items():
                print(f"{nama} - Stok: {info['stok']} - Harga: Rp{info['harga']:,}")
        elif pilihan == "2":
            nama_pembeli = input("Masukkan nama Anda: ")
            print("--- Daftar Sepatu ---")
            for i, (nama, info) in enumerate(stok_sepatu.items(), start=1):
                print(f"{i}. {nama} - Stok: {info['stok']} - Harga: Rp{info['harga']:,}")
            pilihan_sepatu = input("Masukkan nama sepatu yang ingin dipesan: ")
            if pilihan_sepatu in stok_sepatu:
                jumlah = int(input("Jumlah yang ingin dipesan: "))
                if jumlah <= stok_sepatu[pilihan_sepatu]['stok']:
                    total_harga = jumlah * stok_sepatu[pilihan_sepatu]['harga']
                    print(f"Total yang harus dibayar: Rp{total_harga:,}")
                    stok_sepatu[pilihan_sepatu]['stok'] -= jumlah
                    riwayat_pembelian.append({
                        "nama_pembeli": nama_pembeli,
                        "nama_sepatu": pilihan_sepatu,
                        "jumlah": jumlah,
                        "total_harga": total_harga
                    })
                    data_pembeli.append(nama_pembeli)
                    print("Pesanan berhasil!")
                else:
                    print("Stok tidak mencukupi.")
            else:
                print("Sepatu tidak ditemukan.")
        elif pilihan == "0":
            break
        else:
            print("Pilihan tidak valid.")

# Main Loop
while True:
    print("\n=== Sistem Toko Sepatu ===")
    print("1. Masuk sebagai Penjual")
    print("2. Masuk sebagai Pembeli")
    print("0. Keluar")

    menu = input("Pilih peran: ")
    if menu == "1":
        menu_penjual()
    elif menu == "2":
        menu_pembeli()
    elif menu == "0":
        print("Terima kasih telah menggunakan sistem ini.")
        break
    else:
        print("Pilihan tidak valid.")
```
